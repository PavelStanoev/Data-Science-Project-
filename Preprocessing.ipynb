```python
import pandas as pd
```

```python
raw_csv_data = pd.read_csv('Absenteeism_data.csv')
```

```python
df = raw_csv_data.copy()
```

```python
pd.options.display.max_columns = None
pd.options.display.max_rows = None
```

Drop ID ( not needed in our analysis)

```python
df = df.drop(['ID'], axis = 1)
```

Preprocessing Reasons for Absence Columns

```python
df["Reason for Absence"]
```

```python
pd.unique(df["Reason for Absence"])
```

```python
len(df["Reason for Absence"].unique())
```

```python
sorted(df["Reason for Absence"].unique())
```

```python
reason = pd.get_dummies(df["Reason for Absence"])
```

Add a column to check if workers were absent for more than one reason

```python
reason['check if more than one reason'] = reason.sum(axis = 1)
```

If valid drop the newly made column

```python
reason = reason.drop(['check if more than one reason'], axis = 1)
```

Droping column with reason 0 to avoid potential multicollinearity issues

```python
reason = pd.get_dummies(df["Reason for Absence"], drop_first = True)
```

Group the reasons for absence

Drop the Reason for Absence column and classify the dummy variable , so there arent many columns. Class 1 (1-14) - various Diseases ; Class 2 ( 15-17) -  Pregnancy reasons ; Class 3(18-21) - Poisoning types ; Class 4 (22-28) - Light reasons

```python
reason.columns = reason.columns.astype(int)
```

```python
reason_type_1 = reason.loc[:,1:14].max(axis=1)
reason_type_2 = reason.loc[:,15:17].max(axis=1)
reason_type_3 = reason.loc[:,18:21].max(axis=1)
reason_type_4 = reason.loc[:,22:].max(axis=1)
df = df.drop(['Reason for Absence'],axis = 1)
```

Concatenate 

```python
df = pd.concat([df,reason_type_1,reason_type_2,reason_type_3,reason_type_4], axis = 1)
```

```python
df.columns.values
```

```python
columns_names = ['Date', 'Transportation Expense',
       'Distance to Work', 'Age', 'Daily Work Load Average',
       'Body Mass Index', 'Education', 'Children', 'Pets',
       'Absenteeism Time in Hours', 'Reason_1', 'Reason_2', 'Reason_3', 'Reason_4']
```

```python
df.columns = columns_names
```

```python
df.head()
```

Reoder Columns

```python
column_names_new = ['Reason_1', 'Reason_2', 'Reason_3', 'Reason_4','Date', 'Transportation Expense',
       'Distance to Work', 'Age', 'Daily Work Load Average',
       'Body Mass Index', 'Education', 'Children', 'Pets',
       'Absenteeism Time in Hours']
```

```python
df = df[column_names_new]
```

Create a Checkpoint

```python
df_reason_mod = df.copy()
```

Date Column

```python
df_reason_mod['Date']
```

```python
df_reason_mod["Date"] = pd.to_datetime(df_reason_mod["Date"], format = '%d/%m/%Y')
```

```python
df_reason_mod["Date"]
```

Extract the month of the date

```python
list_months =[]
```

```python
for i in range(df_reason_mod.shape[0]):
    list_months.append(df_reason_mod['Date'][i].month)
```

```python
list_months
```

```python
df_reason_mod['Month Value'] = list_months
```

```python
df_reason_mod.head()
```

Extract the day of the date

```python
def date_to_weekday(date):
    return date.weekday()
```

```python
df_reason_mod['Day of the Week'] = df_reason_mod['Date'].apply(date_to_weekday)
```

```python
df_reason_mod.head()
```

```python
df_reason_mod = df_reason_mod.drop(['Date'], axis = 1)
```

```python
df_reason_mod.columns.values
```

```python
df_reason_mod_new_columns = ['Reason_1', 'Reason_2', 'Reason_3', 'Reason_4', 'Month Value',
       'Day of the Week','Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education',
       'Children', 'Pets', 'Absenteeism Time in Hours']
```

```python
df_reason_mod = df_reason_mod[df_reason_mod_new_columns]
```

```python
df_reason_date_mod = df_reason_mod.copy()
```
